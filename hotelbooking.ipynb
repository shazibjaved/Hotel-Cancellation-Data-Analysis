DATA ANALYSIS AND VISUALIZATION

```python
cancelled_perc=df['is_canceled'].value_counts(normalize=True)
print(cancelled_perc)
plt.title('Reservation Status')
plt.bar(['Not Canceled', 'Canceled'], df['is_canceled'].value_counts(), edgecolor='k',width=0.7)
plt.show()
```

```python
df.info()
```

```python
plt.figure(figsize=(8,4))
ax=sns.countplot(x='hotel',data=df,hue='is_canceled',palette='Blues')
legend_labels,_=ax.get_legend_handles_labels()
ax.legend(legend_labels, ['Not Canceled', 'Canceled'], bbox_to_anchor=(1, 1))
plt.title('Reservation Status In Diff Hotels')
plt.xlabel('hotel')
plt.ylabel('number of reservation')
plt.show()
```

```python
resort_hotel=df[df['hotel']=='Resort Hotel']
resort_hotel['is_canceled'].value_counts(normalize=True)
```

```python
city_hotel=df[df['hotel']=='City Hotel']
city_hotel['is_canceled'].value_counts(normalize=True)
```

```python
resort_hotel=resort_hotel.groupby('reservation_status_date')[['Average_Daily_Rate']].mean()
city_hotel=city_hotel.groupby('reservation_status_date')[['Average_Daily_Rate']].mean()
```

```python
plt.figure(figsize=(20,8))
plt.title('Average Daily Rate In City And Resort Hotel',fontsize=30)
plt.plot(resort_hotel.index,resort_hotel['Average_Daily_Rate'],label='Resort_Hotel')
plt.plot(city_hotel.index,city_hotel['Average_Daily_Rate'],label='City_Hotel')
plt.legend(fontsize=20)
plt.show()
```

```python
df['month']=df['reservation_status_date'].dt.month
plt.figure(figsize=(20,8))
sns.countplot(x='month',hue='is_canceled',data=df)
plt.title('Reservation status per Month')
plt.xlabel('Months')
plt.ylabel('Number of reservations')
plt.legend(['not_cancelled','Cancelled'])
plt.show()
```

```python
plt.figure(figsize=(15, 8))
df_grouped = df[df['is_canceled'] == 1].groupby(['month', 'is_canceled'], as_index=False)['Average_Daily_Rate'].sum()
sns.barplot(x='month', y='Average_Daily_Rate', hue='is_canceled', data=df_grouped)
plt.title('ADR Per Month', fontsize=30)
plt.xlabel('Month', fontsize=15)
plt.ylabel('Average Daily Rate', fontsize=15)
plt.show()
```

```python
cancelled_data=df[df['is_canceled']==1]
top10_country= cancelled_data['country'].value_counts()[:10]
plt.figure(figsize=(8,8))
plt.title('Top 10 Countries With Highest Cancellations')
plt.pie(top10_country,autopct='%.1f',labels=top10_country.index)
plt.show()
```

```python
cancelled_data['market_segment'].value_counts(normalize=True)
```

```python
cancelled_df_adr=cancelled_data.groupby('reservation_status_date')[['Average_Daily_Rate']].mean()
cancelled_df_adr.reset_index(inplace=True)
cancelled_df_adr.sort_values('reservation_status_date',inplace=True)

not_cancelled_data=df[df['is_canceled']==0]
not_cancelled_df_adr=not_cancelled_data.groupby('reservation_status_date')[['Average_Daily_Rate']].mean()
not_cancelled_df_adr.reset_index(inplace=True)
not_cancelled_df_adr.sort_values('reservation_status_date',inplace=True)

plt.figure(figsize=(20,8))
plt.title('Average Daily Rate')
plt.plot(not_cancelled_df_adr['reservation_status_date'],not_cancelled_df_adr['Average_Daily_Rate'],label='not cancelled')
plt.plot(cancelled_df_adr['reservation_status_date'],cancelled_df_adr['Average_Daily_Rate'],label='cancelled')
plt.legend()
```

```python
cancelled_df_adr=cancelled_df_adr[(cancelled_df_adr['reservation_status_date']>'2016')&(cancelled_df_adr['reservation_status_date']<'2017-09')]
not_cancelled_df_adr=not_cancelled_df_adr[(not_cancelled_df_adr['reservation_status_date']>'2016')&(not_cancelled_df_adr['reservation_status_date']<'2017-09')]
plt.figure(figsize=(20,8))
plt.title('Average Daily Rate',fontsize=30)
plt.plot(not_cancelled_df_adr['reservation_status_date'],not_cancelled_df_adr['Average_Daily_Rate'],label='not cancelled')
plt.plot(cancelled_df_adr['reservation_status_date'],cancelled_df_adr['Average_Daily_Rate'],label='cancelled')
plt.legend()
```
